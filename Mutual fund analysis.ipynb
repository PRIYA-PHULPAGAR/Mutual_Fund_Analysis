Mutual Fund Analysis
 Objective:
 The objective of this project is to analyze the performance of various mutual funds over time, examining key financial metrics such as Net Asset Value (NAV), returns, expense ratios, and risk levels. This analysis helps in understanding fund performance trends, the impact of expense ratios on returns, and risk distribution among funds.

 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/ADMIN/Downloads/mutual_fund_data.csv",parse_dates=["Date"])
```

```python
# Display basic information and statistics
print(df.info())
```

```python
print(df.describe())
```

```python
# Visualize NAV trends for each fund
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Date", y="NAV", hue="Fund")
plt.title("Mutual Fund NAV Trends")
plt.xlabel("Date")
plt.ylabel("Net Asset Value (NAV)")
plt.legend(title="Fund")
plt.show()
```

```python
# Correlation between Expense Ratio and Returns
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Expense Ratio (%)", y="Return (%)", hue="Fund")
plt.title("Expense Ratio vs. Returns")
plt.xlabel("Expense Ratio (%)")
plt.ylabel("Return (%)")
plt.show()
```

```python
# Boxplot for return distribution per fund
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Fund", y="Return (%)")
plt.title("Return Distribution by Fund")
plt.xlabel("Fund")
plt.ylabel("Return (%)")
plt.show()
```

```python
# Risk category distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Risk Level", palette="Set2")
plt.title("Distribution of Funds by Risk Level")
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.show()
```

The analysis includes trends in NAV, correlation between expense ratio and returns, return distributions, and risk category distributions.
